## Merging the datasets:

```python
#we'll use concatenate since we don't have a key column to join off of
df_combined = pd.concat([df_oldstats, df_newstats], axis=1)
df_combined.head(5)
```

Now that we got these datasets combined, let's break down on how we can get the total number of YouTubers for each country in this timeline.

```python
#let's combine the YouTuber and country columns for each pair ('23 YouTuber/'23 Country and '24 YouTuber/'24 Country)
df_combined['Pair_1'] = df_combined["'23 YouTubers"] + '_' + df_combined["'23 Country"]
df_combined['Pair_2'] = df_combined["'24 YouTubers"] + '_' + df_combined["'24 Country"]

#stack pairs and count the occurrences of YouTubers in countries
combined_pairs = pd.concat([df_combined['Pair_1'], df_combined['Pair_2']])
country_counts = combined_pairs.value_counts().reset_index()
country_counts.columns = ['Country_YouTuber_Pair', 'Count']

#now we'll split the combined pairs back into YouTuber and country
country_counts[['YouTuber', 'Country']] = country_counts['Country_YouTuber_Pair'].str.split('_', expand=True)

#count the distinct number of times all YouTubers are associated with each country
distinct_country_counts = country_counts.groupby('Country')['YouTuber'].nunique().reset_index()
distinct_country_counts.columns = ['Country', 'Distinct_YouTuber_Count']

#sort the dataframe in descending order based on the distinct count of associated YouTubers and then by country for alphabetical order
distinct_country_counts_sorted = distinct_country_counts.sort_values(by=['Distinct_YouTuber_Count', 'Country'], ascending=[False, True])

#display our results
distinct_country_counts_sorted
```

## Visualizations:

#### We'll use a multi bar plot to visualize how the subscriber count has changed between '23 and '24 for the top YouTubers:

```python
#defining the data
labels = df_subs_comb["'23 YouTubers"]
subs_23 = df_subs_comb["'23 Subs"]
subs_24 = df_subs_comb["'24 Subs"]

#setting the width of the bars and the label locations
x = np.arange(len(labels))
width = 0.35

#adjusting the figsize as needed
fig, ax = plt.subplots(figsize=(15, 6))
rects1 = ax.bar(x - width/2, subs_23, width, label="'23 Subs")
rects2 = ax.bar(x + width/2, subs_24, width, label="'24 Subs")

#adding the text for labels, title, and custom x-axis tick labels, etc.
ax.set_xlabel("YouTubers")
ax.set_ylabel("Subscribers (Millions)")
ax.set_title("Subscribers by YouTubers ('23 vs '24)")
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=90)
ax.legend()

#removed scientific notation from y-axis
plt.ticklabel_format(style='plain', axis='y')

#tight_layout method resizes the subplots so they don't overlap one another
plt.tight_layout()

plt.show()
```

From this, we can see that MrBeast gained the most subscribers between July 2023 to January 2024, but we're unable to tell the exact amount gained. Let's create a calculation to display the exact number of subscribers each YouTuber gained in the half year period and sort them in descending order.

```python
#define the function
def diff(a, b): 
    return b - a 

#create the new column by taking both subscriber columns as arguments and then using the apply method to apply it to all the values throught the column
df_subs_comb['Total_Subs_Gained'] = df_subs_comb.apply( 
    lambda x: diff(x["'23 Subs"], x["'24 Subs"]), axis=1)

#sorting the new column by using the .sort_values method
total_subs_sort = df_subs_comb.sort_values(by=["Total_Subs_Gained"], ascending=False)

total_subs_sort.head(None)
```
