## Merging the datasets:

```python
#we'll use concatenate since we don't have a key column to join off of
df_combined = pd.concat([df_oldstats, df_newstats], axis=1)
df_combined.head(5)
```

## Create new dataframe with dropped cols:

```python
#this dataframe will cover our subscriber rankings across '23 to '24
df_subs_comb = df_combined.drop(columns = ["'23 Rank", 
                                           "'23 Uploads", 
                                           "'23 Video Views", 
                                           "'23 Category", 
                                           "'23 Country", 
                                           "'23 Pop.", 
                                           "'24 Rank",
                                           "'24 Avg. Views", 
                                           "'24 Avg. Likes", 
                                           "'24 Avg. Comments", 
                                           "'24 Category", 
                                           "'24 Country"])
df_subs_comb.head(5)
```

## Create new dataframe of subscribers:

```python
#this dataframe will include the top 50 YouTubers
#commented out the code below to avoid creating another subs_comb_stats csv file
#df_subs_comb.to_csv('comb_datasets/subs_comb_stats.csv', index=False)
```

```python
df_subs_comb.drop(df_subs_comb.index[range(10, 50)], axis=0, inplace=True)

df_subs_comb.head(None)
```

```python
#now we'll create a list of the top ten subs and youtubers of both 2023 and 2024
old_subs = df_subs_comb["'23 Subs"].tolist()
new_subs = df_subs_comb["'24 Subs"].tolist()
old_youtubers = df_subs_comb["'23 YouTubers"].tolist()
new_youtubers = df_subs_comb["'24 YouTubers"].tolist()

print(old_subs)
print(new_subs)
print(old_youtubers)
print(new_youtubers)
```

```python
#define dictionary of '23 and '24 YouTubers and their subscriber counts
subs_in_year = {
    "old_youtubers":[old_subs], 
    "new_youtubers": [new_subs]
}

print(subs_in_year)
```

#### We'll use a multi bar plot to visualize how the subscriber count has changed between '23 and '24 for the top ten YouTubers:

```python
#make a data definition
#sub_data = {"'23 YouTubers": [old_subs],
        #"'24 YouTubers": [new_subs]}

#create new dataframe of dictionary with youtubers as the list
bar_chart_df = pd.DataFrame(subs_in_year,columns=["'23 YouTubers", "'24 YouTubers"], index = [old_youtubers, new_youtubers])

#multiple bar chart
bar_chart_df.plot.bar()

#display the plot
plt.show()
```
